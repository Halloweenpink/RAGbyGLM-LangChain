```python
# 填写您自己的APIKey
from langchain_community.chat_models import ChatZhipuAI

ZHIPUAI_API_KEY = "xxx"
llm = ChatZhipuAI(
    temperature=0.1,
    api_key=ZHIPUAI_API_KEY,
    model_name="glm-4",
)
```

```python
from langchain_community.embeddings import ZhipuAIEmbeddings

embeddings = ZhipuAIEmbeddings(
    model="embedding-3",
    api_key=ZHIPUAI_API_KEY,


)
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader

loader_eng = CSVLoader(file_path=r'DouBan.csv')
data_dou = loader_eng.load()
```

```python
data_dou
```

```python
from langchain_community.vectorstores import FAISS
vector = FAISS.from_documents(data_dou, embeddings)
```

```python
from langchain_community.vectorstores import FAISS
vector = FAISS.from_documents([data_dou[0]], embeddings)

for d in data_dou[1:]:
    vector.merge_from(FAISS.from_documents([d], embeddings))

vector.docstore._dict
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template("""仅根据所提供的上下文回答以下问题:

<context>
{context}
</context>

问题: {question}""")

from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
retriever_chain = (
    {"context": retriever , "question": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
retriever = vector.as_retriever(search_type="similarity", search_kwargs={"k": 10})
input_text = '周杰伦的歌千里之外和费玉清合作的时候，费玉清刚进演播室，几小时就完成了？'
retriever.invoke(input_text)
```

```python
retriever_chain.invoke(input_text)
```
